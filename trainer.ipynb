```python
import os
import json
def load_data(data_path):
        # 有进度条展示
        #if self.data_size <= 5:
        data = {}
        a=1
        for a_file in os.listdir(data_path):
            # for mac osx
            if a_file.startswith('.'):
                continue

            
            with open(os.path.join(data_path,a_file), "r", encoding='utf8') as f:
                lines = json.load(f)
                for line in lines:
                    print(line)
                    data.update(line)

        #if self.data_size > 5:
        #    data_gen.close()
        return data

load_data('files/dataset/wyl/')
```
