```python

import argparse
from utils import HParam
hp = HParam('config/test_bert_gpt2.yaml')
```

```python

from gpt2_generator import gpt2_model_gpt2_generator, GPT2_BaseLitModel
import os
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained(hp.tokenizer.tokenizer_path)

gpt2_litmodel = GPT2_BaseLitModel
gpt2_model = gpt2_model_gpt2_generator(hp.gpt2.pretrained_file)
for ckpt_file in os.listdir('files/gpt2_gpt2_final_model/'):
    if ckpt_file.endswith('.ckpt'):
        break

ckpt_path = os.path.join(hp.gpt2.pretrained_file, ckpt_file)
gpt2_litmodel = gpt2_litmodel.load_from_checkpoint(ckpt_path, model=gpt2_model)
```

```python
text=['蓝方直升机被红方载具击伤撤退，多名步兵由D6绕至D7',]
inputs = tokenizer.encode_plus(text,
                                        max_length = 1024,
                                        padding='max_length',
                                        add_special_tokens=True,
                                        return_attention_mask=True,
                                        return_tensors="pt")
```

```python
print(inputs)
logits=gpt2_litmodel.predict(inputs)

sentens = tokenizer.decode(logits)

print(sentens)
```
