```python
import torch
from transformers import GPT2LMHeadModel
BASE_MODEL = 'IDEA-CCNL/Ziya-LLaMA-13B-v1.1'
cache_dir = '/root/autodl-tmp/ziya'
model = LlamaForCausalLM.from_pretrained(
    BASE_MODEL,
    torch_dtype=torch.float16,
    cache_dir = cache_dir,
)
 
tokenizer = LlamaTokenizer.from_pretrained(BASE_MODEL)
```

```python
tokenized_input = tokenizer.encode_plus(['large','small'])
print(tokenized_input)
input_ids = torch.tensor([tokenized_input['input_ids']])
attention_mask = torch.tensor([tokenized_input['attention_mask']])

out = model(input_ids,attention_mask)
```
